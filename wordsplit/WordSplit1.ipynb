# 机械分词

```python
#正向最小
def forwardMin(s,dic,maxL):
    res_dic=[]
    i=0
    length=2
    while i<(len(s)):
        if s[i:i+length] in dic:
            res_dic.append(s[i:i+length])
            i+=length
            length=2
        elif length+1<=maxL:
            length+=1
        else:
            res_dic.append(s[i])
            i+=1
            length=2
    return res_dic
```

```python
#正向最大
def forwardMax(s,dic,maxL):
    res_dic=[]
    i=0
    length=maxL
    while i<(len(s)):
        if s[i:min(len(s),i+length)] in dic:
            res_dic.append(s[i:min(len(s),i+length)])
            i+=length
            length=maxL
        elif length-1>=2:
            length-=1
        else:
            res_dic.append(s[i])
            i+=1
            length=maxL
    return res_dic
```

```python
#逆向最小
def backwardMin(s,dic,maxL):
    res_dic=[]
    i=len(s)-1
    length=2
    while i>=0:
        if s[i-length+1:i+1] in dic:
            res_dic.insert(0,s[i-length+1:i+1])
            i-=length
            length=2
        elif length+1<=maxL:
            length+=1
        else:
            res_dic.insert(0,s[i])
            i-=1
            length=2
    return res_dic
```

```python
#逆向最大
def backwardMax(s,dic,maxL):
    res_dic=[]
    i=len(s)-1
    length=maxL
    while i>=0:
        if s[max(0,i-length+1):i+1] in dic:
            res_dic.insert(0,s[max(0,i-length+1):i+1])
            i-=length
            length=maxL
        elif length-1>=2:
            length-=1
        else:
            res_dic.insert(0,s[i])
            i-=1
            length=maxL
    return res_dic
```

```python
#load dictionary
dic=[]
maxL=0
for i in range(2):
    f = open('wordlist'+str(i)+'.txt','r', encoding='UTF-8')
    if f:
        line = f.readline()
        while line:
            if len(line)-1>maxL:
                maxL=len(line)-1
            dic.append(line[:-1])
            line = f.readline()
        f.close()
    else:
        print('Open file error!')
print(maxL)
print(len(dic))
print(dic[-100:-1])
```

```python
s='我有80个朋友，我不喜欢第2个'
print(forwardMin(s,dic,maxL))
print(forwardMax(s,dic,maxL))
print(backwardMin(s,dic,maxL))
print(backwardMax(s,dic,maxL))
```
