# 自分割分词

## 改良自分割
先将string s按标点符号、英语、数字分割，结果存入list s_list;  
length from n to 2:  
----清空list s1_list;  
----将s_list中的每个string s1取出：  
--------每个s1[i:i+length]计数+1;  
--------将所有计数>=2的词则将其加入dic;  
----将s_list中的每个string s1取出：  
--------如果s1[0]是标点符号则直接将s1加入list s1_list;  
--------如果s1是dic中的词则直接将s1加入s1_list;  
--------如果s1的长度小于length则直接将s1加入s1_list;  
--------否则对s1进行分词:  
------------res_s1清空  
------------对s1进行遍历:  
----------------如果s1[i:i+length]在dic中则先把string temp append到list res_s1中并清空temp，再将其append到list res_s1中;  
----------------否则将s[i]append到string temp中;  
------------将未分割完的s1部分加入到temp中;  
------------将temp append到res_s1中;  
------------将res_s1中的所有string依次放入list s1_list;  
----s_list=s1_list;  
return s_list;  

```python
punc_list=['，','。','“','”','：']
```

```python
def buildCount(s,n):
    #count times that each substring occurs
    count={}
    for length in range(n,1,-1):
        for i in range(len(s)-length+1):
            if s[i:i+length] in count.keys():
                count[s[i:i+length]]+=1
            else:
                count[s[i:i+length]]=1
    return count
```

```python
def autoSplit(s,n,split_to_single=False):
    s_list=[]       #result
    temp=''
    dic=[]
    count={}
    
    #split num, alphabet, puctuation
    i=0
    while i<len(s):
        if s[i] in punc_list:               #punctuation
            if len(temp)>0:
                s_list.append(temp)
                temp=''
            s_list.append(s[i])
            i+=1
        elif s[i]>='0' and s[i]<='9':       #num
            s_list.append(temp)
            temp=''
            while s[i]>='0' and s[i]<='9':
                temp+=s[i]
                i+=1
            s_list.append(temp)
            dic.append(temp)
            temp=''
        elif s[i]>='A' and s[i]<='Z' or s[i]>='a' and s[i]<='z':       #alphabet
            s_list.append(temp)
            temp=''
            while s[i]>='A' and s[i]<='Z' or s[i]>='a' and s[i]<='z':
                temp+=s[i]
                i+=1
            s_list.append(temp)
            dic.append(temp)
            temp=''
        else:                                #other character
            temp+=s[i]
            i+=1
    if len(temp)>0:
        s_list.append(temp)
    
    #split subsentence
    for length in range(n,1,-1):
        s1_list=[]
        for s1 in s_list:
            if s1 in punc_list or s1 in dic or len(s1)<length:
                continue
            else:
                for i in range(len(s1)-length+1):
                    if s1[i:i+length] in count.keys():
                        count[s1[i:i+length]]+=1
                    else:
                        count[s1[i:i+length]]=1
                for key in count.keys():
                    if count[key]>=2:
                        dic.append(key)
        for s1 in s_list:
            if s1 in punc_list or s1 in dic or len(s1)<length:
                s1_list.append(s1)
            else:
                #split s1
                temp=''
                res_s1=[]
                i=0
                while i<len(s1)-length+1:
                    if s1[i:i+length] in dic:
                        if len(temp)>0:
                            res_s1.append(temp)
                            temp=''
                        res_s1.append(s1[i:i+length])
                        i+=length
                    else:
                        temp+=s1[i]
                        i+=1
                while i<len(s1):
                    temp+=s1[i]
                    i+=1
                if len(temp)>0:
                    res_s1.append(temp)
                for s2 in res_s1:
                    if split_to_single and length==2 and len(s2)>2:
                        #if still longer than 2, then we split it to every single character
                        for i in range(len(s2)):
                            s1_list.append(s2[i])
                    else:
                        s1_list.append(s2)
        s_list=s1_list[:]
    return s_list
```

```python
s='英国外相斯特劳对BBC说，英国可能将会在2006年举办公民投票'
print(autoSplit(s,4))
```

```python
s='明天天气很好，明天可以在好天气下出去玩'
print(autoSplit(s,4))
```

```python
s='我有80个朋友，我不喜欢第2个朋友'
print(autoSplit(s,4))
```
